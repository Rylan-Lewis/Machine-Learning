```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import _tree
import re
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
```

### (a) Obtain Data

Get the Accute Inamations Data Set

```python
base_path = r"../data/"

file = os.path.join(base_path, "diagnosis.data")

cols = [
    "temperature",
    "nausea",
    "lumbar_pain",
    "urine_pushing",
    "micturition_pains",
    "burning_urethra",
    "bladder_inflammation",
    "nephritis"
]

acute_df = pd.read_csv(file, header=None, names=cols, sep="\t", engine="python", encoding="utf-16")

# Converted the temperature attribute to numeric by replacing the comma with decimal point
acute_df["temperature"] = acute_df["temperature"].astype(str).str.replace(",", ".").astype(float)

print("Shape:", acute_df.shape)
print("\nFirst few rows:")
print(acute_df.head())
```

### (b) Build a decision tree

```python
# converting the yes and no to 1s and 0s for the model
acute_df_encoded = acute_df.copy()
acute_df_encoded = acute_df_encoded.map(lambda x: 1 if str(x).lower() == "yes" else (0 if str(x).lower() == "no" else x))

# making the dependent and independent variables dataframes
acute_X = acute_df_encoded.drop(columns=["bladder_inflammation", "nephritis"])
acute_y = acute_df_encoded[["bladder_inflammation", "nephritis"]]

# making the decision tree
acute_tree_clf = DecisionTreeClassifier(criterion="gini", random_state=42)
acute_tree_clf.fit(acute_X, acute_y)

plt.figure(figsize=(20, 10))
plot_tree(
    acute_tree_clf,
    feature_names=acute_X.columns,
    class_names=["No", "Yes"],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Multi-Label Decision Tree for predicting the acute inflammation or acute nephritesis of the urinary bladder", fontsize=16)
plt.show()
```

### (c) Convert the decision rules

In the IF part, it combines all the conditions (path_conditions) that led to the current leaf using AND. and creates a complete logical rule like IF (feature1  x) AND (feature2 > y). If there were no splits in a root-only tree, it sets IF TRUE to show that the rule always applies.


In the THEN part, if its a classification tree, it looks at the class counts in the current leaf (value[node][0]),and picks the class with the highest count using np.argmax(), and assigns its label.

```python
# making the function to convert the tree to IF-THEN rules
def tree_to_rules(tree, feature_names, class_names=None):
    tree_ = tree.tree_
    feature = tree_.feature
    threshold = tree_.threshold
    children_left = tree_.children_left
    children_right = tree_.children_right
    value = tree_.value 
    n_node_samples = tree_.n_node_samples
    
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in feature
    ]

    rules = []

    # This recursive function traverses the decision tree, building IFTHEN rules by recording feature-based split conditions along each path, 
    # and when it reaches a leaf node, it makes the complete rule showing the conditions  predicted class and the number of samples in that leaf.
    def recurse(node, path_conditions):
        # for the intermediate nodes
        if children_left[node] != _tree.TREE_LEAF:
            # If feature <= threshold then add it to the left branch of the tree
            cond = f"({feature_name[node]} <= {threshold[node]:.4f})"
            recurse(children_left[node], path_conditions + [cond])
            # If feature > threshold then add it to the right branch of the tree
            cond = f"({feature_name[node]} > {threshold[node]:.4f})"
            recurse(children_right[node], path_conditions + [cond])
        else:
            # on reaching leaf node, the condition building happens here. 
            if path_conditions:
                if_clause = " AND ".join(path_conditions)
            else:
                if_clause = "TRUE"
            if class_names is not None:
                counts = value[node][0]
                class_idx = np.argmax(counts)
                class_label = class_names[class_idx]
                then_clause = f"class = {class_label}"
            else:
                then_clause = f"value = {value[node].tolist()}"            
            n_samples = n_node_samples[node]

            #final rule making
            rule = f"IF {if_clause} THEN {then_clause}  [samples: {n_samples}]"
            rules.append(rule)
    
    recurse(0, [])
    return rules
```

```python
rules = tree_to_rules(acute_tree_clf, feature_names=acute_X.columns)
for rule in rules:
    print(rule)
```

Temperature appears as the root condition, so its the most influential feature separating the data at threshold 37.9500.

Lumbar pain, nausea, and urine pushing appear in deeper splits, acting as more intricate conditions for prediction for more specific subgroups at thresholds of 0.5000 each.

The rules sort patients into simple, easy-to-understand groups. For example:

        1. Low temperature and low urine pushing likely mean no inflammation for both conditions.
        
        2. High temperature with lumbar pain and nausea suggests a higher chance of inflammation.

### (d) Cost-Complexity Pruning to find a minimal decision tree

```python
# here i calculated cost-complexity pruning path
path = acute_tree_clf.cost_complexity_pruning_path(acute_X, acute_y)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

plt.figure(figsize=(8, 5))
plt.plot(ccp_alphas[:-1], impurities[:-1], marker='o', drawstyle="steps-post")
plt.xlabel("Effective alpha")
plt.ylabel("Total impurity of leaves")
plt.title("Cost-Complexity Pruning Path")
plt.show()

# cross validation to select best alpha value
cv = KFold(n_splits=5, shuffle=True, random_state=42)

alpha_scores = []
for alpha in ccp_alphas[:-1]:
    clf = DecisionTreeClassifier(
        random_state=42,
        criterion="gini",
        ccp_alpha=alpha
    )
    scores = cross_val_score(clf, acute_X, acute_y, cv=cv, scoring='f1_micro')
    alpha_scores.append(scores.mean())

alpha_scores = np.array(alpha_scores)
best_alpha = ccp_alphas[:-1][np.argmax(alpha_scores)]

print(f"Best alpha: {best_alpha:.5f}")
print(f"Best CV FI micro score: {alpha_scores.max():.4f}")

plt.figure(figsize=(8, 5))
plt.plot(ccp_alphas[:-1], alpha_scores, marker='o', drawstyle="steps-post")
plt.xlabel("ccp_alpha")
plt.ylabel("Mean CV F1 micro score")
plt.title("Cross-Validation Performance vs Alpha")
plt.show()

# train the final tree on the best alpha
pruned_tree = DecisionTreeClassifier(
    criterion="gini",
    random_state=42,
    ccp_alpha=best_alpha
)
pruned_tree.fit(acute_X, acute_y)

plt.figure(figsize=(18, 9))
plot_tree(
    pruned_tree,
    feature_names=acute_X.columns,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title(f"Pruned Decision Tree", fontsize=16)
plt.show()
```

```python
for alpha in sorted(set(ccp_alphas)):
    clf = DecisionTreeClassifier(ccp_alpha=alpha, random_state=42)
    clf.fit(acute_X, acute_y)
    print(f"alpha={alpha:.4f}, nodes={clf.tree_.node_count}, depth={clf.tree_.max_depth}")
```

```python
rules_pruned = tree_to_rules(pruned_tree, feature_names=acute_X.columns)

print("\nIFTHEN Rules from Pruned Tree\n")
for rule in rules_pruned:
    print(rule)
```

The cross-validation method chose the best ccp_alpha of 0.0, which can also be observed based one the node-depth balance in the pruning path. As ccp_alpha increases, the tree will become simpler and the number of nodes lessens from 11 to 1, leading to the tree depth decreasing from 4 to 0. However, the model accuracy will also deteriorate because the tree begins to miss important decision conditions. Since the cost complexity pruning path and the cross-validation outcome both suggest that the unpruned tree (for ccp_alpha of 0.0) has the best performance, this indicates that the tree has already achieved the best bias-variance trade-off. Thus, the model has not overfit excessively. Hence, pruning in this instance will not be required and hence the pruned tree and the pruned rule base are the exact copies of the original decision tree that was fit on the data.

## 2. The LASSO and Boosting for Regression

### (a) Obtain Data

```python
crime_data = r"../data/communities.data"
crime_names = r"../data/communities.names"

# In the names file, the features are all identifiable with the '@attribute' prefix, so thats why the following regex is writtent to get the column names.
column_names = []
with open(crime_names, 'r') as f:
    for line in f:
        line = line.strip()
        match = re.match(r"@attribute\s+([^\s]+)", line, re.IGNORECASE)
        if match:
            column_names.append(match.group(1))

crime_df = pd.read_csv(crime_data, header=None, names=column_names)

print("Dataset shape:", crime_df.shape)
print("Features:", crime_df.columns.tolist())
```

### (d) Calculate the Coefficient of Variation CV

Gives an idea about the relative variability of a feature with respect to its mean, wether it is highly dispersed over a wider range or it is balanced around its mean.

```python
s = crime_train_df_imputed.std()
m = crime_train_df_imputed.mean()

# coefficient of variation
cv = s / m

# in case of division by 0
cv = cv.replace([np.inf, -np.inf], np.nan).dropna()

print("Coefficient of Variation(CV) of all features:\n", cv)
```

Features like population (CV  2.24), racepctblack (CV  1.43), racePctAsian (CV  1.36), and LemasPctOfficDrugUn (CV  2.55) show high CV, suggesting that these features differ significantly across communities with respect to the mean value.

Features like householdsize (CV  0.36), racePctWhite (CV  0.33), and LemasPctPolicOnPatr (CV  0.12) show low variability, indicating that their values are relatively stable across the dataset.

### (e) Scatter plots and box plots for highest CV features

```python
top_11_features = cv.sort_values(ascending=False).head(int(np.floor(np.sqrt(128))))
print("Top 11 features with highest CV:")
print(top_11_features)
```

```python
target_col = 'ViolentCrimesPerPop'

# Box plot
plt.figure(figsize=(14, 6))
crime_train_df_imputed[top_11_features.index.tolist()].boxplot(rot=90)
plt.title("Boxplots of Top Features by CV")
plt.ylabel("Value")
plt.show()
```

The boxplots reveal a huge amount of outliers for each of these features, a true indication of data being highly variable(dispersed) around theirmeans. However it could also mean that this variability is being explained by extreme values which may or may not be beneficial data as a lot of it could involve noise too. Therefore, high CV alone does not guarantee that a feature is uniformly informative  some of the variation may be due to rare or extreme observations, which could affect model stability and interpretation."

```python
# Scatter plots
for feat in top_11_features.index.tolist():
    plt.figure(figsize=(5, 4))
    sns.scatterplot(x=crime_train_df_imputed[feat], y=crime_train_df_imputed[target_col])
    plt.title(f"{feat} vs ViolentCrimesPerPop")
    plt.xlabel(feat)
    plt.ylabel("ViolentCrimesPerPop")
    plt.show()
```

The scatterplots are unable to show proper relationship between the features as the data is too dispersed. A scatterplot is anyways not really equipped to give an accurate answer about the significance of the feature as even if it shows some relationship of the independent feature with the dependent target variable, we have no way to know if that relationship is due to some random chance, or due to an interaction with another independent feature that is resulting in the relationship with the dependent variable, etc. In order to understand significance of a feature, more concrete methods would be required for example calculating p values of each feature by fitting the model on the data and checking which fall below the threshold, or finding interaction terms which can give an idea about the true nature of the feature, etc.

### (f) Fit a linear model

```python
X_train = crime_train_df_imputed.drop(columns=[target_col])
y_train = crime_train_df_imputed[target_col]
X_test = crime_test_df_imputed.drop(columns=[target_col])
y_test = crime_test_df_imputed[target_col]
```

```python
# linreg model
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

# predict
y_pred_lin = lin_model.predict(X_test)

#error
lin_mse = mean_squared_error(y_test, y_pred_lin)
lin_rmse = np.sqrt(lin_mse)

print(f"Test MSE of lr: {lin_mse:.4f}")
print(f"Test RMSE of lr: {lin_rmse:.4f}")
```

### (g) Fit a ridge regression model

```python
# listing different lambda values to try
lambda_r = np.logspace(-10, 10, 21)

# ridgecv to perform cross validation to find best lambda value
ridge_cv = RidgeCV(alphas = lambda_r, cv =5)
ridge_cv.fit(X_train, y_train)

print("Ridge best alpha:", ridge_cv.alpha_)

# predict
y_pred_ridge = ridge_cv.predict(X_test)

# error
test_mse_ridge = mean_squared_error(y_test, y_pred_ridge)
test_rmse_ridge = np.sqrt(test_mse_ridge)

print("Test MSE:", test_mse_ridge)
print("Test RMSE:", test_rmse_ridge)
```

### (h) Fit a LASSO model

Since dataset is already normalized, directly use the data without using scaler

```python
#lasso
lasso = LassoCV(alphas = lambda_r, cv=5, random_state=42, max_iter=10000)
lasso.fit(X_train, y_train)

# Predict
y_pred_lasso = lasso.predict(X_test)
test_mse_lasso = mean_squared_error(y_test, y_pred_lasso)
test_rmse_lasso = np.sqrt(test_mse_lasso)

# check which features lasso selected
selected_features = X_train.columns[lasso.coef_ != 0]

print("LASSO best alpha:", lasso.alpha_)
print("Test MSE:", test_mse_lasso)
print("Test RMSE:", test_rmse_lasso)
print("Number of features selected:", len(selected_features))
print("Selected features:\n", selected_features.tolist())
```

The Ridge regression model gives the best performance among the three regression models. Though the difference in performance is not very very large, but penalizing the larger coefficients with a lamda value of 1 proves to improve the model's generalization to test data better than the unregularized LR model. This is reflected in the lower test MSE and RMSE compared to the unregularized model, indicating that Ridge regression better controls overfitting.

### (i) Fit a PCR model

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# no. of components
m_values = np.arange(1, X_train.shape[1] + 1)
cv_errors = []

print(f"M       CV error")
for m in m_values:
    # Pipeline: PCA  Linear Regression
    model = Pipeline([
        ('pca', PCA(n_components=m)),
        ('lr', LinearRegression())
    ])
    
    scores = cross_val_score(model, X_train, y_train, cv=kf, scoring='neg_mean_squared_error')
    mse = -np.mean(scores)
    cv_errors.append(mse)
    print(f"{m}\t{mse}")

# choosing the number of components that work best by reducing the cv error
best_m = m_values[np.argmin(cv_errors)]
print(f"Best number of components (M): {best_m}")
print(f"Minimum CV error: {min(cv_errors):.4f}")
```

```python
# Train the final PCR model with the best number of components
best_pcr = Pipeline([
    ('pca', PCA(n_components=best_m)),
    ('lr', LinearRegression())
])

best_pcr.fit(X_train, y_train)

# Predict
y_pred_pcr = best_pcr.predict(X_test)

# error
test_mse_pcr = mean_squared_error(y_test, y_pred_pcr)
test_rmse_pcr = np.sqrt(test_mse_pcr)

print(f"PCR Test Error (MSE): {test_mse_pcr:.4f}")
print(f"PCR Test Error (RMSE): {test_rmse_pcr:.4f}")
```

The Principal Component Regression (PCR) model indicates that using 45 principal components minimizes the cross-validation error.This clearly indicates that a large percentage of the variation in the Crime and Communities dataset can be explained by multiple components. The CV error stabilizes after this point. This means that adding more components does not significantly improve generalization. On the test set, the PCR model achieves a low MSE of 0.0183 and an RMSE of 0.1353. This demonstrates good predictive performance. It shows that PCR effectively reduces the dimensionality of the data while keeping the most informative features. As a result, it leads to better generalization compared to fitting a full linear regression model on all the original variables.

### (j) Fit a boosting tree

```python
kf_xg = KFold(n_splits=5, shuffle=True, random_state=42)

cv_results = []

print(f"Alpha   CV Error")
for alpha in lambda_r:
    fold_mses = []
    for train_idx, val_idx in kf_xg.split(X_train):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train[train_idx], y_train[val_idx]

        model_xg = XGBRegressor(
            objective="reg:squarederror",
            reg_alpha=alpha,   
            # reg_lambda=1.0,      
            random_state=42,
            n_jobs=-1
        )

        model_xg.fit(X_tr, y_tr)
        preds_xg = model_xg.predict(X_val)
        mse_xg = mean_squared_error(y_val, preds_xg)
        fold_mses.append(mse_xg)

    avg_mse_xg = np.mean(fold_mses)
    cv_results.append((alpha, avg_mse_xg))
    print(f"{alpha}\t{avg_mse_xg}")

# best alpha
best_alpha_xg, best_mse_xg = min(cv_results, key=lambda x: x[1])
print("\nBest alpha selected by CV:", best_alpha_xg)
print("Best mean CV MSE:", best_mse_xg)

# train model using best alpha found
final_model_xg = XGBRegressor(
    objective="reg:squarederror",
    reg_alpha=best_alpha_xg,
    # reg_lambda=1.0,
    random_state=42
)
final_model_xg.fit(X_train, y_train)

#predict
y_pred_xg = final_model_xg.predict(X_test)
test_mse_xg = mean_squared_error(y_test, y_pred_xg)
test_rmse_xg = np.sqrt(test_mse_xg)

print("Test MSE XGBoost:", test_mse_xg)
print("Test RMSE XGBoost:", test_rmse_xg)
```
